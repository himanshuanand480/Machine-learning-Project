```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv("/content/medicine.csv")
data
```

```python
data['Description']
```

```python
data['Description'].apply(lambda x : x.split())
```

```python
data['Description'] = data['Description'].apply(lambda x : x.split())
data['Reason'] = data['Reason'].apply(lambda x : x.split())
```

```python
data.tail()
```

```python

data['Description'] = data['Description'].apply(lambda x : [i.replace(" ", "") for i in x])
```

```python
data['tags'] = data['Description'] + data['Reason']
```

```python

data.head()
```

```python
new_data = data[['index', 'Drug_Name', 'tags']]
```

```python
new_data.head()
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x : " ".join(x))
```

```python
new_data['tags'] = new_data['tags'].apply(lambda x : x.lower())
```

```python
new_data.head()
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(stop_words = "english", max_features = 5000)
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem("You are so so beautiful")
```

```python
new_data['tags'] = new_data['tags'].apply(stem)
```

```python
new_data.head()
```

```python
vectors = cv.fit_transform(new_data['tags']).toarray()
```

```python
vectors.shape
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommendation(medicine):
    medicine_index = new_data[new_data['Drug_Name'] == medicine].index[0]
    distance = similarity[medicine_index]
    medicine_list = sorted(list(enumerate(distance)), reverse=True, key = lambda x : x[1])[1:6]
    for i in medicine_list:
        print(new_data.iloc[i[0]].Drug_Name)
```

```python
recommendation("ACGEL CL NANO Gel 15gm")
```

```python
# Step 1: Import libraries
import pandas as pd
import pickle
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Step 2: Load medicine data
# If you're using a CSV file:
# medicines = pd.read_csv('your_medicine_data.csv')

# OR if you're loading an old pickle:
# medicines_dict = pickle.load(open('medicine_dict.pkl', 'rb'))
# medicines = pd.DataFrame(medicines_dict)

# ðŸ‘‡ Sample data for demonstration (replace with real data)
medicines = pd.DataFrame({
    'Drug_Name': ['Paracetamol', 'Ibuprofen', 'Aspirin', 'Cetirizine', 'Dolo'],
    'Description': [
        'Used to treat pain and fever.',
        'Nonsteroidal anti-inflammatory drug.',
        'Relieves pain, fever, inflammation.',
        'Used for allergies and cold symptoms.',
        'Pain relief and fever reducer.'
    ]
})

# Step 3: Add image paths assuming youâ€™ll use local images in PyCharm
medicines['Image_URL'] = medicines['Drug_Name'].apply(lambda name: f"images/{name}.jpg")

# Step 4: Compute similarity using TF-IDF on description
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(medicines['Description'])
similarity = cosine_similarity(tfidf_matrix)

# Step 5: Save data for Streamlit
with open('medicine_dict.pkl', 'wb') as f:
    pickle.dump(medicines.to_dict(), f)

with open('similarity.pkl', 'wb') as f:
    pickle.dump(similarity, f)

print("âœ… Files saved: medicine_dict.pkl and similarity.pkl")
```

```python
import pandas as pd
import os

image_folder = '/content/unzipped_data/Medicine_data'  # Path to image folder
medicine_df = pd.read_csv('/content/medicine_data.csv')  # Replace with actual CSV file name

image_mapping = []

for index, row in medicine_df.iterrows():
    drug_name = row['Drug_Name']
    drug_name_clean = drug_name.strip().replace(" ", "_").replace("/", "_")
    image_filename = f"{drug_name_clean}.jpg"
    image_path = os.path.join(image_folder, image_filename)

    if os.path.exists(image_path):
        image_mapping.append(image_path)
    else:
        print(f"Image not found for: {drug_name}")
        image_mapping.append(None)

medicine_df['Image_URL'] = image_mapping
```
